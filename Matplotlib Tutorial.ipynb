# Scatter plots

Scatter plots are used to plot data points on a horizontal and a vertical axis in the attempt to
show how much one variable is affected by another. Each row in the data table is represented by a marker
whose position depends on its values in the columns set on the X and Y axes.

```python
#sample scatter plot
x = np.arange(10)
y = np.random.randint(20,30,10)
plt.scatter(x,y,color = "red")
plt.show()
```

```python
age = np.array([21,24,26,28,30])
sal = np.array([11000,13000, 17000, 24000,20000])
plt.scatter(age,sal,color="orange",marker='*')

plt.xlabel('age')
plt.ylabel('sal')
plt.title(" My Scatter Plot")
plt.show()
```

# Bar Graphs

Bar graphs are used to compare things between different groups or to track changes over time. However, when trying to measure change over time, bar graphs are best when the changes are larger.


A bar chart or bar graph is a chart or graph that presents categorical data with rectangular bars with heights or lengths proportional to the values that they represent. The bars can be plotted vertically or horizontally. A vertical bar chart is sometimes called a column chart

```python
#bar graph with id's and salaries
import pandas as pd
empdata = {"empid"  : [1001,1002,1003,1004,1005,1006],
                      "ename":["Sachin","Srinu","Kumar","Gaurava", "Anil", "Kapil"],
              "sal":[ 10000,23000,34000,45000,27000,61000],
              }
```

```python
#create a dataframe
df = pd.DataFrame(empdata)
```

```python
#extracting empid and sal into x and y variables
x=df['empid']
y=df['sal']
```

```python
x
```

```python
y
```

```python
#create bar graph - v1
plt.bar(x,y, label = 'Employee Data', color ='red')
```

```python
#create bar graph - v2
plt.bar(x,y, label = 'Employee Data', color ='purple')
plt.xlabel("Employee Ids")
plt.ylabel("Employee Salaries")
plt.title("XYZ Pvt Ltd")
plt.legend()
plt.show()
```

# Histogram

A histogram is a graphical display of data using bars of different heights. In a histogram, each bar groups numbers into ranges. Taller bars show that more data falls in that range. A histogram displays the shape and spread of continuous sample data.

A histogram is an approximate representation of the distribution of numerical data

```python
#sample histogram of employee ages
#let us take individual employee ages and range of ages
emp_ages = [22,45,30,59,58,56,57,45,43,50,40,34,33,25,19]
bins = [0,10,20,30,40,50,60]
```

```python
emp_ages
```

```python
bins
```

```python
#create a histogram 
plt.hist(emp_ages,bins, label = 'Emp Data', histtype = 'bar', rwidth = 0.8, color = 'purple')
#set labels
plt.xlabel('employee ages')
plt.ylabel('no. of employees')
plt.title("TITANIC Group")
plt.legend()
plt.show()
```

# Pie Chart

A pie chart is a circular statistical graphic, which is divided into slices to illustrate numerical proportion

Pie charts are generally used to show percentage or proportional data and usually the percentage represented by each category is provided next to the corresponding slice of pie. Pie charts are good for displaying data for around 6 categories or fewer.

```python
#lets display different departments in pie chart
#percentage of employees of a 4 departments
slices = [50,20,15,15]
#take the department names
depts = ['sales','production','HR','Finance']
#take the colors for each department
cols = ['purple','cyan','brown','gold']
#creating a pie chart
plt.pie(slices, labels = depts, colors = cols, startangle=90, explode =( 0,0,0,0), shadow = True, autopct = '%.1f%%')
#set title and legend
plt.title("RELIANCE")
plt.legend()
plt.show()
```
